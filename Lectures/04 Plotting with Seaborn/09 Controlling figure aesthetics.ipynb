# Controlling Figure Aesthetics

Data visualization is not just about representing numbers; it's about telling a compelling story with your data. Seaborn, a powerful statistical data visualization library built on top of Matplotlib, provides an intuitive interface to create attractive and informative graphics. In this lecture, we'll explore how to control figure aesthetics in Seaborn to create visually appealing and effective plots.


🔑 **Key Concept:** Seaborn's aesthetic controls allow you to create professional-looking visualizations with minimal effort, enhancing both data exploration and presentation.


The visual appeal of your plots goes beyond mere decoration:

1. **Clarity**: Well-designed plots make it easier for viewers to understand the data at a glance.
2. **Engagement**: Attractive visualizations capture and maintain the audience's attention.
3. **Professionalism**: Polished graphics lend credibility to your analysis and presentations.


Seaborn takes a unique approach to controlling figure aesthetics:

1. **Built-in Themes**: Seaborn comes with predefined themes that set a cohesive look across all your plots.
2. **Separation of Concerns**: Aesthetics are divided into three main categories:
   - Style: Controls the visual aspects like colors and grids.
   - Context: Manages the scale of plot elements for different output mediums.
   - Palette: Determines the color scheme used in the plots.
3. **Matplotlib Integration**: Seaborn builds upon Matplotlib, allowing for further customization when needed.


Seaborn provides four main functions for controlling aesthetics:

1. `set_style()`: Sets the aesthetic style of the plots
2. `set_context()`: Controls the scaling of plot elements
3. `set_palette()`: Sets the color palette for the plots
4. `set_theme()`: Sets the overall theme, encompassing style, context, and palette


Each of these functions has a corresponding function that returns a dictionary of parameters instead of setting them:

- `axes_style()`: Returns style parameters
- `plotting_context()`: Returns context parameters
- `color_palette()`: Returns a list of colors


Let's start with a simple example to demonstrate how Seaborn's default aesthetics differ from Matplotlib's:


````python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
````

````python
# Function to create a simple plot
def sinplot(n=10, flip=1):
    x = np.linspace(0, 14, 100)
    for i in range(1, n + 1):
        plt.plot(x, np.sin(x + i * 0.5) * (n + 2 - i) * flip)
````

````python
# Matplotlib default
plt.subplot(121)
sinplot()
plt.title("Matplotlib Default")

# Seaborn default
sns.set_theme()
plt.subplot(122)
sinplot()
plt.title("Seaborn Default")
````

💡 **Pro Tip:** Always call `sns.set_theme()` at the beginning of your script to ensure consistent Seaborn styling across all your plots.


Seaborn's aesthetic controls offer several advantages:

1. **Consistency**: Apply a uniform look across all your visualizations with minimal code.
2. **Flexibility**: Easily switch between different styles to suit various purposes or preferences.
3. **Readability**: Seaborn's default styles are designed to enhance data readability and interpretation.


🤔 **Why This Matters:** By mastering Seaborn's aesthetic controls, you'll be able to create visually appealing plots that effectively communicate your data insights, whether you're exploring data for yourself or presenting findings to others.


❗️ **Important Note:** While Seaborn provides excellent defaults and easy-to-use aesthetic controls, remember that the most effective visualizations are those that best represent your specific data and communicate your intended message. Always consider your audience and the purpose of your visualization when making aesthetic choices.


In the following sections, we'll dive deeper into how to customize these aesthetics, from setting styles and contexts to choosing color palettes and applying overall themes. By the end of this lecture, you'll have the tools to create beautiful, informative visualizations that make your data shine.

## <a id='toc1_'></a>[Customizing Plot Styles using `set_style()`](#toc0_)

Seaborn offers a powerful way to customize the overall look of your plots using the `set_style()` function. This function controls various elements of the plot's appearance, such as the background color, grid lines, and axis properties.


The `set_style()` function allows you to choose from five preset styles:

1. `"darkgrid"` (default)
2. `"whitegrid"`
3. `"dark"`
4. `"white"`
5. `"ticks"`


🔑 **Key Concept:** Each style is designed for different purposes and can significantly change the look and feel of your plots.


Let's explore these styles with an example:


````python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt


def sinplot(flip=1):
    x = np.linspace(0, 14, 100)
    for i in range(1, 7):
        plt.plot(x, np.sin(x + i * 0.5) * (7 - i) * flip)
    plt.show()


styles = ["darkgrid", "whitegrid", "dark", "white", "ticks"]

for style in styles:
    sns.set_style(style)
    sinplot()
    plt.title(f"Style: {style}")
````

### <a id='toc1_1_'></a>[Style Elements](#toc0_)


When you set a style, you're adjusting several plot elements simultaneously:

- Background color
- Grid presence and style
- Axis line properties
- Tick mark properties


💡 **Pro Tip:** The `"darkgrid"` and `"whitegrid"` styles are excellent for data-heavy plots where precise value lookups are important. The `"dark"` and `"white"` styles are better for plots where overall patterns are more important than individual values.


### <a id='toc1_2_'></a>[Customizing Styles](#toc0_)


You can further customize styles by passing a dictionary of parameters to `set_style()`:


````python
sns.set_style("darkgrid", {"axes.facecolor": ".9", "grid.color": ".6"})
sinplot()
plt.title("Customized Dark Grid")
````

### <a id='toc1_3_'></a>[Temporary Style with `axes_style()`](#toc0_)


The `axes_style()` function is closely related to `set_style()`. It returns a dictionary of parameters that can be used to create a specific style:


````python
style_dict = sns.axes_style("darkgrid")
print(style_dict)
````

You can use `axes_style()` with a context manager to apply a style to a specific part of your code:


````python
with sns.axes_style("darkgrid", {"axes.facecolor": "black", "grid.color": ".6"}):
    plt.figure()
    sinplot()
    plt.title("Temporary White Grid Style")
````

🤔 **Why This Matters:** Using `axes_style()` allows you to create plots with different styles within the same figure, which can be useful for emphasizing different aspects of your data.


### <a id='toc1_4_'></a>[Removing Spines](#toc0_)


Seaborn provides the `despine()` function to remove top and right spines, which are often unnecessary:


````python
sns.set_style("white")
sinplot()
sns.despine(left=True)
plt.title("Plot with Spines Removed")
````

❗️ **Important Note:** While customizing styles can make your plots more attractive, it's crucial to ensure that style choices don't interfere with data interpretation. Always prioritize clarity and accuracy in your visualizations.


By mastering `set_style()` and its related functions, you can create plots that are not only informative but also visually appealing and suited to your specific needs. In the next section, we'll explore how to control the scale of plot elements using `set_context()`.

## <a id='toc2_'></a>[Scaling Plot Elements using `set_context()`](#toc0_)

After setting the style of your plots, you may want to adjust the size and scale of various elements to suit different presentation contexts. Seaborn's `set_context()` function allows you to control the scaling of plot elements, making it easy to create figures that look great in various settings, from academic papers to presentations.


The `set_context()` function allows you to choose from four preset contexts:

1. `"paper"` (smallest elements)
2. `"notebook"` (default)
3. `"talk"` 
4. `"poster"` (largest elements)


🔑 **Key Concept:** These contexts adjust the size of labels, lines, and other plot elements to suit different output mediums.


Let's explore these contexts with an example:


````python
import seaborn as sns
import matplotlib.pyplot as plt

contexts = ["paper", "notebook", "talk", "poster"]

for context in contexts:
    sns.set_context(context)
    sinplot()
    plt.title(f"Context: {context}")
````

### <a id='toc2_1_'></a>[Context Parameters](#toc0_)


When you set a context, you're adjusting several plot parameters:

- Font sizes
- Line widths
- Tick label sizes
- Legend sizes


💡 **Pro Tip:** Use `"paper"` or `"notebook"` for publications and general data exploration, `"talk"` for presentations, and `"poster"` for large-format posters.


### <a id='toc2_2_'></a>[Fine-tuning Context](#toc0_)


You can further customize the context by passing additional parameters:


````python
sns.set_context("talk", rc={"lines.linewidth": 8}, font_scale=2)
sinplot()
plt.title("Customized Talk Context")
````

Here, `font_scale` allows you to adjust font sizes independently of other elements.


### <a id='toc2_3_'></a>[Temporary Context with `plotting_context()`](#toc0_)


Similar to `axes_style()`, the `plotting_context()` function returns a dictionary of parameters for a given context:


```python
context_dict = sns.plotting_context("poster")
print(context_dict)
```


You can use this with a context manager for temporary changes:


````python
with sns.plotting_context("talk", rc={"lines.linewidth": 1}, font_scale=1.5):
    sinplot()
    plt.title("Temporary Paper Context")
````

🤔 **Why This Matters:** Using `plotting_context()` allows you to create plots with different scales within the same script, which can be useful when preparing figures for multiple output formats.


### <a id='toc2_4_'></a>[Combining Style and Context](#toc0_)


For a comprehensive approach, you can combine `set_style()` and `set_context()`:


````python
sns.set_style("whitegrid", {"axes.facecolor": "k"})
sns.set_context("poster")
sinplot()
plt.title("Whitegrid Style with Talk Context")
````

❗️ **Important Note:** While scaling elements can improve readability, be cautious not to oversize elements to the point where they distort data perception. Always test your plots at the intended output size.


### <a id='toc2_5_'></a>[Practical Application](#toc0_)


Consider this real-world scenario:


````python
# Prepare data
np.random.seed(0)
data = np.random.randn(100, 3)

# Create plot for a paper
sns.reset_defaults()
with sns.plotting_context("paper"):
    sns.scatterplot(x=data[:, 0], y=data[:, 1], hue=data[:, 2])
    plt.title("Scatter Plot (Paper Context)")
````

````python
# Create plot for a presentation
with sns.plotting_context("talk"):
    sns.scatterplot(x=data[:, 0], y=data[:, 1], hue=data[:, 2])
    plt.title("Scatter Plot (Talk Context)")
    plt.show()
````

This example demonstrates how to create versions of the same plot optimized for different contexts.


By mastering `set_context()` and its related functions, you can ensure your plots are appropriately scaled for their intended use, whether in a scientific paper, a presentation, or a poster. In the next section, we'll explore how to control color palettes to further enhance your visualizations.

## <a id='toc3_'></a>[Choosing Color Palettes](#toc0_)

Color is a crucial aspect of data visualization. It can highlight important patterns, distinguish between categories, and even evoke emotional responses. Seaborn provides powerful tools for controlling color palettes through the `set_palette()` function and its related `color_palette()` function. Choosing the right color palette can significantly enhance the readability and impact of your visualizations.


Colors can be defined using three components:

1. **Hue**: The property that distinguishes "different colors" (e.g., red, blue).
2. **Saturation**: The colorfulness or intensity of the color.
3. **Luminance**: How much light is emitted, ranging from black to white.


The following guidelines are useful:
1. **Vary hue to distinguish categories**: Use different hues for categorical data.
2. **Vary luminance to represent numbers**: Use changes in brightness for numeric data.


Use hue variations for categorical data and luminance variations for numeric data to create more intuitive and easily interpretable visualizations.


To choose a palette, Seaborn classifies color palettes into three main categories:

1. **Qualitative palettes**: Best for representing categorical data.
2. **Sequential palettes**: Suitable for representing numeric data with a natural ordering.
3. **Diverging palettes**: Ideal for numeric data with a meaningful midpoint.


### <a id='toc3_1_'></a>[Using `color_palette()` and `set_palette()`](#toc0_)


The `color_palette()` function is the primary tool for working with color palettes in Seaborn:


````python
import seaborn as sns
import matplotlib.pyplot as plt

# Get the current default color palette
current_palette = sns.color_palette()
current_palette
````

````python
# Set a new default palette
sns.set_palette("husl")

# Use a specific palette for a plot
sinplot()
````

### <a id='toc3_5_'></a>[Practical Application: Choosing Palettes for Different Data Types](#toc0_)


Let's apply these principles to real data:


````python
import numpy as np
import pandas as pd

# Generate sample data
np.random.seed(0)
data = pd.DataFrame(
    {
        "category": ["A", "B", "C", "D"] * 25,
        "value1": np.random.randn(100),
        "value2": np.random.rand(100),
    }
)
````

## <a id='toc4_'></a>[Setting the Overall Theme](#toc0_)

The `set_theme()` function in Seaborn provides a powerful way to control the overall aesthetic of your plots. This function allows you to set multiple aesthetic parameters at once, including style, context, palette, and font scale. It is a high-level function that combines the effects of `set_style()`, `set_context()`, and `set_palette()` into a single call.


The basic syntax of `set_theme()` is:


```python
sns.set_theme(style=None, context=None, palette=None, font="sans-serif", font_scale=1, color_codes=True, rc=None)
```


Let's break down each parameter:

1. `style`: Overall plot style ("darkgrid", "whitegrid", "dark", "white", "ticks")
2. `context`: Plot element scaling ("paper", "notebook", "talk", "poster")
3. `palette`: Color palette (e.g., "deep", "muted", "bright", "pastel", "dark")
4. `font`: Font family for text elements
5. `font_scale`: Scaling factor for font sizes
6. `color_codes`: Whether to use short color codes for seaborn palette colors
7. `rc`: Dictionary of rc parameter mappings to override


Here's a simple example of how to use `set_theme()`:


````python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Set the overall theme
sns.set_theme(style="whitegrid", context="talk", palette="deep", font_scale=1.2)

sinplot()
````

### <a id='toc4_1_'></a>[Exploring Different Themes](#toc0_)


Let's explore how different theme settings affect the same plot:


````python
def plot_sine_wave():
    x = np.linspace(0, 10, 100)
    y = np.sin(x)
    plt.figure(figsize=(10, 6))
    plt.plot(x, y)
    plt.title("Sine Wave")
    plt.xlabel("X-axis")
    plt.ylabel("Y-axis")
    plt.show()
````

````python
# Default theme
sns.set_theme()
sinplot()
````

````python
# Dark theme with larger font for presentations
sns.set_theme(style="darkgrid", context="talk", palette="bright", font_scale=1.5)
sinplot()
````

````python
# Minimalist theme for papers
sns.set_theme(style="ticks", context="paper", palette="muted", font_scale=0.8)
sinplot()
````

### <a id='toc4_2_'></a>[Customizing RC Parameters](#toc0_)


For more fine-grained control, you can use the `rc` parameter to override specific Matplotlib rc parameters:


````python
sns.set_theme(
    style="whitegrid",
    rc={
        "axes.edgecolor": "0.15",
        "grid.color": "0.85",
        "grid.linestyle": "--",
        "lines.linewidth": 5,
    },
)
sinplot()
````

### <a id='toc4_3_'></a>[Practical Application: Comparing Datasets with Different Themes](#toc0_)


Let's create a more complex example to showcase how different themes can affect data visualization:


````python
import pandas as pd

# Generate sample data
np.random.seed(0)
data = pd.DataFrame(
    {
        "Group A": np.random.normal(0, 1, 1000),
        "Group B": np.random.normal(0.5, 1.5, 1000),
        "Group C": np.random.normal(-0.5, 2, 1000),
    }
)


def plot_distributions(theme_name, **theme_kwargs):
    sns.set_theme(**theme_kwargs)
    sns.histplot(data, kde=True)
    plt.title(f"Distribution Comparison ({theme_name} Theme)")
    plt.xlabel("Value")
    plt.ylabel("Count")
    plt.show()
````

````python
sns.reset_defaults()

# Default theme
plot_distributions("Default")
````

````python
# Dark theme for presentations
plot_distributions(
    "Dark Presentation",
    style="darkgrid",
    context="talk",
    palette="bright",
    font_scale=1.2,
)
````

````python
# Light theme for papers
plot_distributions(
    "Light Paper", style="whitegrid", context="paper", palette="pastel", font_scale=0.9
)
````

````python
# Custom theme
plot_distributions(
    "Custom",
    style="ticks",
    context="notebook",
    palette="deep",
    font_scale=1.1,
    rc={
        "axes.titlesize": 20,
        "axes.labelsize": 14,
        "xtick.labelsize": 12,
        "ytick.labelsize": 12,
    },
)
````

💡 **Pro Tip:** When preparing visualizations for different mediums (e.g., papers, presentations, posters), use `set_theme()` to quickly adjust your plots' appearance to suit the context.


🤔 **Why This Matters:** Consistent theming across your visualizations not only makes them more aesthetically pleasing but also helps in creating a cohesive narrative in your data story. It can significantly improve the readability and impact of your plots, especially when presenting to different audiences or in various formats.


While `set_theme()` provides a convenient way to set multiple aesthetic parameters at once, remember that you can always override specific elements in individual plots if needed. This flexibility allows you to maintain overall consistency while fine-tuning specific visualizations. By mastering `set_theme()`, you can efficiently control the look and feel of your Seaborn plots, ensuring they are not only informative but also visually appealing and appropriate for their intended audience and medium.
