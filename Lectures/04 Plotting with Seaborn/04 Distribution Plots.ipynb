# Distribution Plots in Seaborn

Distribution plots are powerful tools for visualizing the spread and characteristics of data. They provide insights into the underlying patterns, central tendencies, and variability of datasets. In this lecture, we'll explore various distribution plots available in Seaborn, a statistical data visualization library built on top of Matplotlib.


Understanding the distribution of your data is crucial for:
1. Identifying patterns and trends
2. Detecting outliers and anomalies
3. Informing statistical analysis and modeling decisions
4. Communicating findings effectively to stakeholders


Distribution plots help answer key questions about your data, such as:
- What is the central tendency of the data?
- How spread out are the values?
- Is the distribution symmetric or skewed?
- Are there any unusual patterns or multiple modes?


Seaborn offers several types of distribution plots, each with its own strengths:

1. **Histograms**: Show the frequency distribution of a dataset.
2. **Kernel Density Estimation (KDE) Plots**: Provide a smooth, continuous estimation of the probability density function.
3. **Empirical Cumulative Distribution Function (ECDF) Plots**: Display the proportion of data points below each value.
4. **Rug Plots**: Show the individual data points along an axis.
5. **Box Plots**: Summarize the distribution with quartiles and outliers.
6. **Violin Plots**: Combine KDE with box plot features.


Seaborn provides both figure-level and axes-level functions for creating distribution plots:

1. Figure-level functions:
   - `displot()`: Flexible function for plotting univariate or bivariate distributions.
   - `jointplot()`: Plot bivariate distributions with marginal univariate distributions.
   - `pairplot()`: Create a grid of plots for pairwise relationships in a dataset.

2. Axes-level functions:
   - `histplot()`: Create histograms.
   - `kdeplot()`: Create KDE plots.
   - `ecdfplot()`: Create ECDF plots.
   - `rugplot()`: Add a rug plot to an existing axes.

Let's start with a simple example using the `penguins` dataset:


```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
penguins = sns.load_dataset("penguins")

# Create a histogram
sns.displot(data=penguins, x="flipper_length_mm", kde=True)
plt.title("Distribution of Penguin Flipper Lengths")
```

This code will produce a histogram of penguin flipper lengths with an overlaid KDE curve.


Seaborn offers extensive customization options for distribution plots:

- Adjust bin width or number of bins in histograms
- Modify KDE bandwidth
- Change color palettes and styles
- Add conditional distributions using the `hue` parameter
- Customize plot elements like titles, labels, and legends


💡 **Pro Tip:** The choice of distribution plot depends on your data and analysis goals:

- Use histograms for an intuitive view of data frequency
- KDE plots are great for smooth, continuous distributions
- ECDF plots are useful for comparing multiple distributions
- Rug plots can be added to other plots to show individual data points
- Box and violin plots are excellent for comparing distributions across categories


❗️ **Important Note:** Always consider the nature of your data (continuous, discrete, categorical) when choosing a distribution plot type.

## <a id='toc1_'></a>[Histograms with Seaborn](#toc0_)

Histograms are one of the most fundamental and intuitive tools for visualizing the distribution of data. They provide a quick and effective way to understand the shape, central tendency, and spread of a dataset. In this section, we'll explore how to create and customize histograms using Seaborn, a powerful data visualization library built on top of Matplotlib.


Let's start with a simple example using a dataset of penguin measurements:


```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the penguins dataset
penguins = sns.load_dataset("penguins")

# Create a basic histogram
sns.histplot(data=penguins, x="flipper_length_mm")
plt.title("Distribution of Penguin Flipper Lengths")
```

This code will produce a histogram showing the distribution of flipper lengths among penguins.


Histograms are excellent for:
- Visualizing the distribution of a single continuous variable
- Identifying the shape of a distribution (e.g., normal, skewed, bimodal)
- Detecting outliers and unusual patterns in the data
- Comparing distributions across different groups or categories

However, histograms may not be ideal when:
- Dealing with very small datasets (where individual data points matter more)
- Visualizing discrete or categorical data (bar plots might be more appropriate)
- Comparing multiple distributions with significant overlap (density plots or box plots might be better)
- Precise representation of individual data points is required (consider scatter plots or rug plots)


### <a id='toc1_1_'></a>[Customizing Bin Size and Count](#toc0_)


The choice of bins can significantly impact the insights drawn from a histogram. Seaborn offers several ways to control binning:


```python
# Specify the number of bins
sns.histplot(data=penguins, x="flipper_length_mm", bins=30)
plt.title("Histogram with 30 bins")
```

```python
# Specify bin width
sns.histplot(data=penguins, x="flipper_length_mm", binwidth=5)
plt.title("Histogram with 5mm bin width")
```

💡 **Pro Tip:** Always experiment with different bin sizes to ensure you're not missing important features of your data.


### <a id='toc1_2_'></a>[Adding Kernel Density Estimation (KDE)](#toc0_)


To get a smoother representation of the distribution, we can add a KDE curve:


```python
sns.histplot(data=penguins, x="flipper_length_mm", kde=True)
plt.title("Histogram with KDE")
```

### <a id='toc1_3_'></a>[Comparing Multiple Distributions](#toc0_)


Histograms are great for comparing distributions across categories:


```python
sns.histplot(data=penguins, x="flipper_length_mm", hue="species", element="step")
plt.title("Flipper Length Distribution by Species")
```

### <a id='toc1_4_'></a>[Dealing with Skewed Data](#toc0_)


For heavily skewed distributions, using a logarithmic scale can be helpful:


```python
planets = sns.load_dataset("planets")
sns.histplot(data=planets, x="distance", log_scale=False)
plt.title("Distribution of Planetary Distances")
```

```python
planets = sns.load_dataset("planets")
sns.histplot(data=planets, x="distance", log_scale=True)
plt.title("Distribution of Planetary Distances (Log Scale)")
```

### <a id='toc1_5_'></a>[Creating 2D Histograms](#toc0_)


Seaborn also supports bivariate histograms, which are excellent for exploring relationships between two continuous variables:


```python
sns.histplot(data=penguins, x="bill_length_mm", y="bill_depth_mm", cbar=True)
plt.title("2D Histogram of Bill Dimensions")
```

### <a id='toc1_6_'></a>[Customizing Histogram Appearance](#toc0_)


Seaborn offers various options to customize the appearance of histograms:


```python
sns.histplot(
    data=penguins,
    x="flipper_length_mm",
    hue="species",
    multiple="stack",
    palette="viridis",
    edgecolor=".3",
    linewidth=.5
)
plt.title("Stacked Histogram with Custom Styling")
```

Histograms are versatile tools for understanding data distributions. By mastering Seaborn's `histplot()` function, you can create insightful and visually appealing histograms that effectively communicate the structure of your data.


❗️ **Important Note:** Remember that while histograms are powerful, they're just one tool in the data visualization toolkit. Always consider your data's nature and your analysis goals when choosing the most appropriate visualization method.

## <a id='toc2_'></a>[Kernel Density Estimation (KDE) Plots](#toc0_)

Kernel Density Estimation (KDE) plots are sophisticated tools for visualizing the probability density of a continuous variable. They provide a smooth, continuous estimation of the distribution, offering insights that might be obscured by the discrete nature of histograms.


Let's start with a simple example using a dataset of penguin measurements:


```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the penguins dataset
penguins = sns.load_dataset("penguins")

# Create a basic KDE plot
sns.kdeplot(data=penguins, x="flipper_length_mm")
plt.title("Distribution of Penguin Flipper Lengths")
```

This code will produce a smooth KDE plot showing the distribution of flipper lengths among penguins.


KDE plots are excellent for:
- Visualizing smooth, continuous distributions
- Comparing multiple distributions simultaneously
- Identifying multimodal distributions
- Estimating the probability density at any given point

However, KDE plots may not be ideal when:
- Dealing with discrete or categorical data
- Working with small datasets (where individual data points are crucial)
- Analyzing distributions with hard boundaries (e.g., strictly positive data)
- Precise representation of data frequency is required (histograms might be better)


### <a id='toc2_1_'></a>[Adjusting Bandwidth](#toc0_)


The bandwidth parameter is crucial in KDE plots. It determines the smoothness of the curve:


```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(18, 6))

# Under-smoothed
sns.kdeplot(data=penguins, x="flipper_length_mm", ax=ax1, bw_adjust=0.2)
ax1.set_title("Under-smoothed (bw_adjust=0.2)")

# Default smoothing
sns.kdeplot(data=penguins, x="flipper_length_mm", ax=ax2)
ax2.set_title("Default smoothing")

# Over-smoothed
sns.kdeplot(data=penguins, x="flipper_length_mm", ax=ax3, bw_adjust=2)
ax3.set_title("Over-smoothed (bw_adjust=2)")

plt.tight_layout()
```

💡 **Pro Tip:** Always experiment with different bandwidth adjustments to ensure you're not missing important features or introducing artificial ones.


### <a id='toc2_2_'></a>[Comparing Multiple Distributions](#toc0_)


KDE plots excel at comparing distributions across categories:


```python
sns.kdeplot(data=penguins, x="flipper_length_mm", hue="species", fill=True)
plt.title("Flipper Length Distribution by Species")
```

### <a id='toc2_3_'></a>[Bivariate KDE Plots](#toc0_)


Seaborn can create beautiful bivariate KDE plots to visualize the relationship between two continuous variables:


```python
sns.kdeplot(data=penguins, x="flipper_length_mm", y="bill_length_mm", cmap="YlGnBu", fill=True, cbar=True)
plt.title("Bivariate KDE: Flipper Length vs Bill Length")
```

### <a id='toc2_4_'></a>[Cumulative Distribution Function (CDF)](#toc0_)


KDE plots can also show cumulative distributions:


```python
sns.kdeplot(data=penguins, x="flipper_length_mm", hue="species", cumulative=True)
plt.title("Cumulative Distribution of Flipper Lengths by Species")
```

### <a id='toc2_5_'></a>[Dealing with Bounded Data](#toc0_)


For data with natural boundaries (e.g., always positive), we can use the `cut` parameter:


```python
tips = sns.load_dataset("tips")
sns.kdeplot(data=tips, x="total_bill", cut=0)
plt.title("Distribution of Total Bill (Bounded at 0)")
```

KDE plots are powerful tools for visualizing continuous distributions. They offer smooth, intuitive representations of data that can reveal subtle patterns and trends. By mastering Seaborn's `kdeplot()` function, you can create insightful and visually appealing density plots that effectively communicate the structure of your data.


❗️ **Important Note:** While KDE plots are versatile, remember that they always produce a smooth curve. For discrete or "spiky" data, or when precise frequency information is crucial, consider using histograms or other visualization methods alongside KDE plots.

## <a id='toc3_'></a>[Empirical Cumulative Distribution Function (ECDF) Plots](#toc0_)

Empirical Cumulative Distribution Function (ECDF) plots are powerful tools for visualizing the distribution of data. Unlike histograms or KDE plots, ECDF plots represent each data point directly, offering a comprehensive view of the entire dataset without the need for binning or smoothing.


Let's start with a simple example using the penguins dataset:


```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the penguins dataset
penguins = sns.load_dataset("penguins")

# Create a basic ECDF plot
sns.ecdfplot(data=penguins, x="flipper_length_mm")
plt.title("ECDF of Penguin Flipper Lengths")
plt.xlabel("Flipper Length (mm)")
plt.ylabel("Cumulative Proportion")
```

This code will produce an ECDF plot showing the cumulative distribution of flipper lengths among penguins.


ECDF plots are excellent for:
- Visualizing the entire distribution of a dataset without information loss
- Comparing multiple distributions directly
- Identifying percentiles and quantiles easily
- Detecting subtle differences between distributions

However, ECDF plots may not be ideal when:
- Intuitive understanding of central tendency or variance is required at a glance
- Visualizing multimodal distributions (peaks are less obvious)
- Communicating with audiences unfamiliar with cumulative distributions


### <a id='toc3_1_'></a>[Comparing Multiple Distributions](#toc0_)


ECDF plots excel at comparing distributions across categories:


```python
sns.ecdfplot(data=penguins, x="flipper_length_mm", hue="species")
plt.title("ECDF of Flipper Lengths by Species")
plt.xlabel("Flipper Length (mm)")
plt.ylabel("Cumulative Proportion")
```

### <a id='toc3_2_'></a>[Complementary CDF](#toc0_)


Sometimes, it's useful to plot the complementary CDF (1 - CDF):


```python
sns.ecdfplot(data=penguins, x="flipper_length_mm", hue="species", complementary=True)
plt.title("Complementary CDF of Flipper Lengths by Species")
plt.xlabel("Flipper Length (mm)")
plt.ylabel("1 - Cumulative Proportion")
```

### <a id='toc3_3_'></a>[Changing the Statistic](#toc0_)


By default, ECDF plots show proportions, but we can change this to counts or percentages:


```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Count statistic
sns.ecdfplot(data=penguins, x="flipper_length_mm", hue="species", stat="count", ax=ax1)
ax1.set_title("ECDF with Count Statistic")
ax1.set_ylabel("Cumulative Count")

# Percent statistic
sns.ecdfplot(data=penguins, x="flipper_length_mm", hue="species", stat="percent", ax=ax2)
ax2.set_title("ECDF with Percent Statistic")
ax2.set_ylabel("Cumulative Percent")

plt.tight_layout()
```

### <a id='toc3_4_'></a>[Practical Example: Analyzing Income Distribution](#toc0_)


Let's use a real-world example to showcase the power of ECDF plots. We'll analyze income distribution data:


```python
import numpy as np
import pandas as pd

# Generate synthetic income data
np.random.seed(42)
low_income = np.random.lognormal(mean=10, sigma=0.5, size=1000)
high_income = np.random.lognormal(mean=11, sigma=0.5, size=1000)

# Combine data into a DataFrame
income_data = pd.DataFrame({
    'Income': np.concatenate([low_income, high_income]),
    'Group': ['Low Income']*1000 + ['High Income']*1000
})

# Create the ECDF plot
plt.figure(figsize=(12, 6))
sns.ecdfplot(data=income_data, x="Income", hue="Group")
plt.title("Income Distribution Comparison", fontsize=16)
plt.xlabel("Annual Income ($)", fontsize=12)
plt.ylabel("Cumulative Proportion", fontsize=12)
plt.xscale('log')  # Use log scale for better visualization of income distribution

# Add annotations for median income
for group in ['Low Income', 'High Income']:
    median = income_data[income_data['Group'] == group]['Income'].median()
    plt.axvline(median, linestyle='--', color='gray', alpha=0.7)
    plt.text(median, 0.5, f'{group} Median: ${median:.0f}', 
             rotation=90, verticalalignment='center')

plt.tight_layout()
```

This ECDF plot visualizes the income distribution for two groups, clearly showing the differences in their distributions and allowing for easy comparison of percentiles.


### <a id='toc3_5_'></a>[Advanced ECDF Techniques](#toc0_)


1. **Highlighting Percentiles:**


```python
def highlight_percentile(data, percentile, ax, color):
    value = np.percentile(data, percentile)
    ax.axvline(value, color=color, linestyle='--', alpha=0.7)
    ax.axhline(percentile/100, color=color, linestyle='--', alpha=0.7)
    ax.text(value, 0.02, f'{percentile}th\npercentile', 
            color=color, ha='right', va='bottom', rotation=90)

plt.figure(figsize=(10, 6))
ax = sns.ecdfplot(data=penguins, x="flipper_length_mm")
highlight_percentile(penguins["flipper_length_mm"].dropna(), 25, ax, 'red')
highlight_percentile(penguins["flipper_length_mm"].dropna(), 75, ax, 'green')
plt.title("ECDF of Penguin Flipper Lengths with Quartiles Highlighted")
```

2. **Comparing to Theoretical Distribution:**


```python
from scipy import stats

sns.ecdfplot(data=penguins, x="flipper_length_mm", label="Empirical")

# Generate theoretical normal distribution
mean = penguins["flipper_length_mm"].mean()
std = penguins["flipper_length_mm"].std()
x = np.linspace(mean - 3*std, mean + 3*std, 100)
y = stats.norm.cdf(x, mean, std)
plt.plot(x, y, label="Theoretical Normal", linestyle='--')

plt.title("Empirical vs Theoretical CDF of Penguin Flipper Lengths")
plt.legend()
```

ECDF plots are powerful tools for visualizing and comparing distributions. They offer a direct representation of each data point, making them particularly useful for detecting subtle differences between distributions and identifying percentiles. By mastering Seaborn's `ecdfplot()` function, you can create insightful visualizations that capture the full range of your data without loss of information.


❗️ **Important Note:** While ECDF plots provide a comprehensive view of the data, they may be less intuitive for audiences unfamiliar with cumulative distributions. Consider using them in conjunction with other plots like histograms or box plots for a more complete picture.

## <a id='toc4_'></a>[Rug Plots and Their Applications](#toc0_)

Rug plots are a subtle yet powerful tool in the data visualization toolkit. They provide a unique way to show the distribution of individual data points along one or both axes of a plot, complementing other visualization techniques and offering additional insights into the data structure.


Let's start with a simple example using the tips dataset:


```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the tips dataset
tips = sns.load_dataset("tips")

# Create a basic rug plot
sns.rugplot(data=tips, x="total_bill")
plt.title("Distribution of Total Bill Amounts")
plt.xlabel("Total Bill ($)")
```

This code will produce a rug plot showing the distribution of total bill amounts along the x-axis.


Rug plots are excellent for:
- Showing the exact location of individual data points
- Complementing other plots like scatter plots or KDE plots
- Visualizing the density of data along an axis
- Identifying clusters or gaps in the data distribution

However, rug plots may not be ideal when:
- Dealing with very large datasets (the plot can become cluttered)
- The main focus is on overall distribution shape (histograms or KDE plots might be better)
- Precise quantitative comparisons are needed
- The plot already contains many elements, and adding rugs might overwhelm the viewer


### <a id='toc4_1_'></a>[Combining Rug Plots with Other Visualizations](#toc0_)


Rug plots truly shine when combined with other plots. Let's create a more informative visualization by combining a rug plot with a KDE plot:


```python
sns.kdeplot(data=tips, x="total_bill", fill=True)
sns.rugplot(data=tips, x="total_bill", color="red", alpha=0.5)
plt.title("Distribution of Total Bill Amounts with Rug Plot")
plt.xlabel("Total Bill ($)")
```

### <a id='toc4_2_'></a>[Two-Dimensional Rug Plots](#toc0_)


Rug plots can also be used to show distributions in two dimensions:


```python
sns.scatterplot(data=tips, x="total_bill", y="tip")
sns.rugplot(data=tips, x="total_bill", y="tip", alpha=0.5)
plt.title("Relationship between Total Bill and Tip with Rug Plot")
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")
```

### <a id='toc4_3_'></a>[Customizing Rug Plots](#toc0_)


You can customize rug plots to make them more informative or visually appealing:


```python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="time", style="time")
sns.rugplot(data=tips, x="total_bill", y="tip", hue="time", height=0.05, alpha=0.5)
plt.title("Relationship between Total Bill and Tip by Time of Day")
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")
```

Rug plots are a versatile and informative addition to many types of visualizations. They provide a way to see the individual data points without cluttering the main plot, offering insights into data density, clusters, and gaps. By mastering the use of rug plots in Seaborn, you can create more informative and nuanced visualizations that reveal subtle patterns in your data.


💡 **Pro Tip:** Use rug plots judiciously. They work best when they add information without overwhelming the main visualization. Experiment with alpha values and line widths to find the right balance for your data. While rug plots are powerful, they can become less effective with very large datasets. In such cases, consider using sampling techniques or more advanced density visualization methods.

## <a id='toc5_'></a>[Figure Level Distribution Visualization with `sns.displot()`](#toc0_)

The `sns.displot()` function is a powerful, figure-level interface in Seaborn that provides a versatile approach to visualizing univariate and bivariate distributions. It combines the functionality of `histplot()`, `kdeplot()`, and `ecdfplot()`, allowing for rich, multifaceted visualizations of data distributions.


Let's start with a simple example using the penguins dataset:


```python
import seaborn as sns
import matplotlib.pyplot as plt

penguins = sns.load_dataset("penguins")

# Basic histogram
sns.displot(data=penguins, x="flipper_length_mm")
plt.title("Distribution of Penguin Flipper Lengths")
```

Use `sns.displot()` when you want to:
- Create complex, multi-faceted distribution plots
- Easily switch between different types of distribution plots (histogram, KDE, ECDF)
- Visualize distributions across multiple subgroups or conditions
- Combine different types of plots (e.g., histogram with KDE overlay)


### <a id='toc5_1_'></a>[Switching Between Plot Types](#toc0_)


`sns.displot()` allows easy switching between different types of plots:


```python
# Histogram
sns.displot(data=penguins, x="flipper_length_mm", kind="hist")
```

```python
# KDE plot
sns.displot(data=penguins, x="flipper_length_mm", kind="kde")
```

```python
# ECDF plot
sns.displot(data=penguins, x="flipper_length_mm", kind="ecdf")
```

### <a id='toc5_2_'></a>[Combining Plot Types](#toc0_)


You can combine different plot types for richer visualizations:


```python
sns.displot(data=penguins, x="flipper_length_mm", kde=True, rug=True)
plt.title("Histogram with KDE and Rug Plot")
```

### <a id='toc5_3_'></a>[Bivariate Distributions](#toc0_)


`sns.displot()` can create bivariate distribution plots:


```python
sns.displot(data=penguins, x="flipper_length_mm", y="bill_length_mm", kind="kde")
plt.title("Bivariate KDE: Flipper Length vs Bill Length")
```

### <a id='toc5_4_'></a>[Faceted Plots](#toc0_)


One of the most powerful features of `sns.displot()` is the ability to create faceted plots:


```python
g = sns.displot(
    data=penguins,
    x="flipper_length_mm",
    hue="species",
    col="sex",
    kind="kde",
    height=4,
)
g.figure.suptitle("Flipper Length Distribution by Species and Sex", y=1.02)
```

### <a id='toc5_5_'></a>[Customization and Fine-tuning](#toc0_)


`sns.displot()` returns a `FacetGrid` object, which allows for further customization:


```python
g = sns.displot(
    data=penguins,
    x="flipper_length_mm",
    hue="species",
    col="sex",
    kind="kde",
    height=4,
    aspect=.7,
)
g.set_axis_labels("Flipper Length (mm)", "Density")
g.set_titles("{col_name} Penguins")
g.figure.suptitle("Penguin Flipper Length Distribution", y=1.02)
```

`sns.displot()` is a versatile and powerful tool for creating comprehensive distribution visualizations. It combines the strengths of different plot types and allows for easy creation of complex, multi-faceted visualizations. By mastering `sns.displot()`, you can create insightful, publication-quality distribution plots that effectively communicate the structure and patterns in your data.


💡 **Pro Tip:** Experiment with different combinations of plot types, faceting variables, and aesthetic mappings to find the most effective way to represent your data.


❗️ **Important Note:** While `sns.displot()` offers great flexibility, it's important to choose the representation that best suits your data and analysis goals. Sometimes, simpler plots might be more effective in conveying your message.

## <a id='toc6_'></a>[Visualizing Bivariate Distributions](#toc0_)

Bivariate distributions are a powerful way to explore relationships between two continuous variables. Seaborn provides several methods to visualize these distributions, offering insights into correlation, clustering, and patterns that might not be apparent when examining variables individually. Here are some of the most common types of bivariate distribution plots in Seaborn:
1. Scatter plots
2. 2D Histogram plots
3. 2D Kernel Density Estimation (KDE) plots
4. Hexbin plots
5. Joint plots


Use bivariate distribution plots when you want to:
- Explore relationships between two continuous variables
- Identify patterns, correlations, or clusters in your data
- Detect outliers or unusual data points
- Compare distributions across different subgroups


Let's explore each of these types with examples.


### <a id='toc6_1_'></a>[Scatter Plots](#toc0_)


Scatter plots are the most basic form of bivariate visualization:


```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
tips = sns.load_dataset("tips")

# Create a scatter plot
sns.scatterplot(data=tips, x="total_bill", y="tip")
plt.title("Relationship between Total Bill and Tip")
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")
```

💡 **Pro Tip:** Add a third variable using the `hue` parameter for more insights:


```python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="time")
plt.title("Relationship between Total Bill and Tip, by Time of Day")
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")
```

### <a id='toc6_2_'></a>[2D Histogram Plots](#toc0_)


2D histograms show the frequency of data points in a grid:


```python
sns.histplot(data=tips, x="total_bill", y="tip", bins=20, cbar=True)
plt.title("2D Histogram of Total Bill vs Tip")
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")
```

### <a id='toc6_3_'></a>[2D Kernel Density Estimation (KDE) Plots](#toc0_)


KDE plots provide a smooth estimation of the bivariate probability density function:


```python
sns.kdeplot(data=tips, x="total_bill", y="tip", fill=True, cmap="YlOrRd", cbar=True)
plt.title("2D KDE Plot of Total Bill vs Tip")
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")
```

`resid` option will regress y on x (possibly as a robust or polynomial regression) and then draw a scatterplot of the residuals. You can optionally fit a **lowess (locally weighted scatterplot smoothing)** smoother to the residual plot, which can help in determining if there is structure to the residuals.

```python
x = np.random.normal(0, 1, 100)
y = 4 + 3 * x ** 2 + np.random.normal(0, 2, 100)

sns.jointplot(x=x, y=y, kind="resid", order=1, lowess=True, line_kws={"color": "red"})
```

```python
x = np.random.normal(0, 1, 100)
y = 4 + 3 * x ** 2 + np.random.normal(0, 2, 100)

sns.jointplot(x=x, y=y, kind="resid", order=2, lowess=True)
```

To add more layers onto the plot, use the methods on the JointGrid object that jointplot() returns:

## <a id='toc7_'></a>[Summary](#toc0_)

Throughout this course, we've explored various techniques for visualizing distributions using Seaborn. Let's recap the key points and provide some guidelines for choosing the right plot for your data.

1. **Histograms**:
   - Ideal for visualizing frequency distributions
   - Sensitive to bin size choices
   - Great for identifying modes and general shape of the distribution

2. **Kernel Density Estimation (KDE) Plots**:
   - Provide smooth, continuous estimates of distributions
   - Useful for comparing multiple distributions
   - Bandwidth selection is crucial for accurate representation

3. **Empirical Cumulative Distribution Function (ECDF) Plots**:
   - Represent each data point without binning or smoothing
   - Excellent for comparing distributions and identifying percentiles
   - Less intuitive for audiences unfamiliar with cumulative distributions

4. **Box Plots and Violin Plots**:
   - Summarize distributions with key statistics
   - Useful for comparing distributions across categories
   - Violin plots combine KDE with box plot features

5. **Rug Plots**:
   - Show individual data points along an axis
   - Often used in combination with other distribution plots


The choice of distribution plot depends on your data and your analytical goals:

1. **For a quick overview of the data shape:**
   - Use histograms or KDE plots

2. **When precise representation of each data point matters:**
   - Use ECDF plots or rug plots

3. **For comparing multiple distributions:**
   - Use overlaid KDE plots or ECDF plots

4. **When dealing with categorical variables:**
   - Use box plots or violin plots

5. **For bivariate distributions:**
   - Use 2D histograms or bivariate KDE plots

6. **When you want to show the raw data alongside the distribution:**
   - Combine rug plots with KDE or histogram plots
