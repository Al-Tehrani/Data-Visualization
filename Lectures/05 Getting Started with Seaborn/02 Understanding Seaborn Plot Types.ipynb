# Understanding Seaborn Plot Types

Welcome to our second lecture on Seaborn! In this session, we'll dive deeper into understanding the various plot types offered by Seaborn and how they can be effectively used to visualize different kinds of data.


Seaborn provides a rich set of plotting functions that cater to a wide range of data visualization needs. These plot types can be broadly categorized into several groups:

1. **Relational plots**: Used to visualize the relationship between two or more variables.
   - Examples: scatterplot, lineplot

2. **Categorical plots**: Ideal for comparing distributions across categories.
   - Examples: boxplot, violinplot, barplot

3. **Distribution plots**: Help in understanding the distribution of data.
   - Examples: histplot, kdeplot, ecdfplot

4. **Regression plots**: Useful for showing the relationship between variables along with a fitted regression model.
   - Examples: regplot, lmplot

5. **Matrix plots**: Excellent for visualizing complex datasets in a matrix format.
   - Examples: heatmap, clustermap


Let's look at a quick example that showcases a few of these plot types:


````python
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline

# Load a sample dataset
tips = sns.load_dataset("tips")

# Set up the matplotlib figure
fig, axes = plt.subplots(2, 2, figsize=(8, 6))

# Relational plot: scatterplot
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="time", ax=axes[0, 0])
axes[0, 0].set_title("Scatterplot: Relational")

# Categorical plot: boxplot
sns.boxplot(data=tips, x="day", y="total_bill", ax=axes[0, 1])
axes[0, 1].set_title("Boxplot: Categorical")

# Distribution plot: histplot
sns.histplot(data=tips, x="total_bill", kde=True, ax=axes[1, 0])
axes[1, 0].set_title("Histplot: Distribution")

# Regression plot: regplot
sns.regplot(data=tips, x="total_bill", y="tip", ax=axes[1, 1])
axes[1, 1].set_title("Regplot: Regression")

plt.tight_layout()
````

This example demonstrates four different types of Seaborn plots, each serving a distinct purpose:

1. The scatterplot shows the relationship between total bill and tip, with different colors for lunch and dinner.
2. The boxplot compares the distribution of total bill across different days of the week.
3. The histplot displays the distribution of total bills, including a kernel density estimate.
4. The regplot shows the relationship between total bill and tip, including a linear regression line.


As we progress through this lecture, we'll explore how these different plot types work, when to use them, and how to customize them to best represent your data. We'll also delve into the distinction between figure-level and axes-level functions, which is crucial for understanding how to create both simple and complex visualizations with Seaborn.

## <a id='toc1_'></a>[How Different Plots in Seaborn Work](#toc0_)

Seaborn provides a consistent and intuitive API for creating various types of statistical plots. Understanding the basic principles of how these plots work will enable you to create and customize visualizations effectively.


### <a id='toc1_1_'></a>[Basic principles of Seaborn's API](#toc0_)


1. **Data-aware plotting**: Seaborn functions are designed to work directly with Pandas DataFrames. They can automatically extract and plot variables based on column names.

2. **Declarative API**: You specify what you want to plot, and Seaborn handles the details of creating the visualization.

3. **Sensible defaults**: Seaborn comes with aesthetically pleasing default styles and color palettes, making it easy to create attractive plots quickly.

4. **Statistical estimation**: Many Seaborn plots automatically include statistical information like confidence intervals.


Let's explore these principles with some examples:


````python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the tips dataset
tips = sns.load_dataset("tips")

# Create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(8, 4))

# Create a simple scatterplot
sns.scatterplot(data=tips, x="total_bill", y="tip", ax=ax1)
ax1.set_title("Basic Scatterplot")

# Create a more complex plot with additional variables
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="time", size="size", ax=ax2)
ax2.set_title("Scatterplot with Multiple Variables")

plt.tight_layout()
````

In these examples, you can see how Seaborn automatically handles axis labels, legend creation, and color mapping based on the data provided.


### <a id='toc1_2_'></a>[Common parameters and customization options](#toc0_)


Most Seaborn plotting functions share a set of common parameters that allow for easy customization:

1. **data**: The DataFrame containing the dataset.
2. **x, y**: Names of variables in the DataFrame for the x and y axes.
3. **hue**: Variable for color encoding.
4. **size**: Variable for size encoding of plot elements.
5. **style**: Variable for different marker styles.
6. **palette**: Color palette to use.
7. **ax**: The Matplotlib axis to draw the plot on.


Here's an example demonstrating some of these customization options:


````python
# Create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# Customized scatterplot
sns.scatterplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="time",
    size="size",
    style="day",
    palette="deep",
    sizes=(20, 200),
    ax=ax1
)
ax1.set_title("Highly Customized Scatterplot")

# Categorical plot with customization
sns.boxplot(
    data=tips,
    x="day",
    y="total_bill",
    hue="time",
    palette="pastel",
    width=0.6,
    ax=ax2
)
ax2.set_title("Customized Boxplot")

plt.tight_layout()
````

These examples showcase how you can use various parameters to add more dimensions to your plots and customize their appearance.


### <a id='toc1_3_'></a>[How different plots work](#toc0_)


Different plot types in Seaborn are designed to highlight specific aspects of your data:

1. **Relational plots** (e.g., scatterplot, lineplot): Show the relationship between two or more variables. 
2. **Categorical plots** (e.g., boxplot, violinplot): Display the distribution of a quantitative variable across categories.
3. **Distribution plots** (e.g., histplot, kdeplot): Visualize the distribution of a single variable.
4. **Regression plots** (e.g., regplot, lmplot): Combine relational plots with statistical model fitting.
5. **Matrix plots** (e.g., heatmap, clustermap): Represent data in a grid format.


Here's an example comparing different plot types:


````python
# Set up a 2x2 grid of subplots
fig, axes = plt.subplots(2, 2, figsize=(8, 8))

# Relational plot
sns.scatterplot(data=tips, x="total_bill", y="tip", ax=axes[0, 0])
axes[0, 0].set_title("Scatterplot (Relational)")

# Categorical plot
sns.boxplot(data=tips, x="day", y="total_bill", ax=axes[0, 1])
axes[0, 1].set_title("Boxplot (Categorical)")

# Distribution plot
sns.histplot(data=tips, x="total_bill", kde=True, ax=axes[1, 0])
axes[1, 0].set_title("Histplot (Distribution)")

# Regression plot
sns.regplot(data=tips, x="total_bill", y="tip", ax=axes[1, 1])
axes[1, 1].set_title("Regplot (Regression)")

plt.tight_layout()
````

This comparison demonstrates how different plot types can reveal various aspects of the same dataset. Understanding these differences will help you choose the most appropriate plot type for your specific data and analysis goals.


By consistently using the `ax` parameter in Seaborn functions and working with subplots, we maintain better control over our figure layout and adhere to best practices in combining Seaborn with Matplotlib.

## <a id='toc2_'></a>[Figure-level vs. Axes-level functions](#toc0_)

Understanding the distinction between figure-level and axes-level functions in Seaborn is crucial for creating both simple and complex visualizations effectively. This knowledge allows you to choose the right approach based on your specific visualization needs.


### <a id='toc2_1_'></a>[Explanation of the difference](#toc0_)


1. **Axes-level functions:**
   - Work on a single Matplotlib Axes object.
   - Allow for more precise control over plot placement and integration with existing Matplotlib figures.
   - Generally start with lower-case letters (e.g., `scatterplot`, `boxplot`).

2. **Figure-level functions:**
   - Create a new figure and set of axes, often with multiple subplots.
   - Provide a higher-level interface for certain types of plots.
   - Generally start with capital letters (e.g., `FacetGrid`, `PairGrid`) or are specific plot types (e.g., `lmplot`, `relplot`).


### <a id='toc2_2_'></a>[When to use each type](#toc0_)


- Use **axes-level functions** when:
  - You want to create a single plot or add a plot to an existing figure.
  - You need fine-grained control over plot placement.
  - You're combining multiple plot types in one figure.

- Use **figure-level functions** when:
  - You want to create a standalone, complex figure with minimal code.
  - You need to create multiple related plots (e.g., faceted plots).
  - You want to take advantage of Seaborn's automatic figure management.


### <a id='toc2_3_'></a>[Examples of figure-level functions](#toc0_)


Let's look at some examples of figure-level functions:


````python
import seaborn as sns
import matplotlib.pyplot as plt

# Load example datasets
tips = sns.load_dataset("tips")
iris = sns.load_dataset("iris")
````

````python
# Example 1: lmplot (figure-level)
sns.lmplot(data=tips, x="total_bill", y="tip", hue="time", height=4)
plt.title("lmplot: Regression plot with faceting")
````

````python
# Example 2: relplot (figure-level)
sns.relplot(data=iris, x="sepal_length", y="petal_length", hue="species", kind="scatter", height=4)
plt.title("relplot: Scatter plot with multiple semantics")
plt.show()
````

````python
# Example 3: FacetGrid (figure-level)
g = sns.FacetGrid(tips, col="time", row="smoker", height=3)
g.map(sns.scatterplot, "total_bill", "tip")
g.add_legend()
````

### <a id='toc2_4_'></a>[Examples of axes-level functions](#toc0_)


Now, let's compare with axes-level functions:


````python
# Create a 2x2 grid of subplots
fig, axes = plt.subplots(2, 2, figsize=(8, 8))

# Example 1: scatterplot (axes-level)
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="time", ax=axes[0, 0])
axes[0, 0].set_title("scatterplot: Basic scatter plot")

# Example 2: boxplot (axes-level)
sns.boxplot(data=tips, x="day", y="total_bill", ax=axes[0, 1])
axes[0, 1].set_title("boxplot: Box plot for categorical data")

# Example 3: histplot (axes-level)
sns.histplot(data=tips, x="total_bill", kde=True, ax=axes[1, 0])
axes[1, 0].set_title("histplot: Histogram with KDE")

# Example 4: regplot (axes-level)
sns.regplot(data=tips, x="total_bill", y="tip", ax=axes[1, 1])
axes[1, 1].set_title("regplot: Scatter plot with regression line")

plt.tight_layout()
````

### <a id='toc2_5_'></a>[Key Differences](#toc0_)


1. **Plot Creation:**
   - Axes-level functions add to existing axes or create a single plot.
   - Figure-level functions often create new figures with one or more subplots.

2. **Customization:**
   - Axes-level functions offer more direct control over individual plot elements.
   - Figure-level functions provide higher-level parameters for customizing multiple related plots at once.

3. **Complexity:**
   - Axes-level functions are simpler and more flexible for basic plots.
   - Figure-level functions are powerful for creating complex, multi-plot visualizations with less code.

4. **Integration:**
   - Axes-level functions integrate more easily with existing Matplotlib code.
   - Figure-level functions are more self-contained and manage figure creation internally.


Understanding these differences allows you to choose the most appropriate function type for your visualization needs, whether you're creating a simple plot or a complex, multi-faceted visualization.

## <a id='toc3_'></a>[Integration with Matplotlib](#toc0_)

Seaborn is built on top of Matplotlib and is designed to integrate seamlessly with it. This integration allows you to leverage the strengths of both libraries, combining Seaborn's statistical plotting capabilities with Matplotlib's fine-grained control over plot elements. Understanding how Seaborn works with Matplotlib is crucial for creating customized and complex visualizations.


### <a id='toc3_1_'></a>[How Seaborn works on top of Matplotlib](#toc0_)


1. **Building on Matplotlib's foundation:** 
   Seaborn uses Matplotlib's plotting functions under the hood, adding statistical and aesthetic enhancements.

2. **Consistent styling:** 
   Seaborn provides a set of themes that can be applied to Matplotlib plots, ensuring consistency across your visualizations.

3. **Higher-level interface:** 
   Seaborn offers a more intuitive API for certain types of statistical plots, simplifying the process of creating complex visualizations.

4. **Compatible objects:** 
   Seaborn functions return Matplotlib Axes or FigureAxes objects, allowing further customization using Matplotlib methods.


### <a id='toc3_2_'></a>[Combining Seaborn and Matplotlib commands](#toc0_)


Let's explore how to combine Seaborn and Matplotlib commands effectively:


````python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Load a sample dataset
tips = sns.load_dataset("tips")

# Create a figure and axes
fig, ax = plt.subplots(figsize=(8, 6))

# Use Seaborn to create a scatter plot
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="time", size="size", ax=ax)

# Use Matplotlib to customize the plot
ax.set_title("Combining Seaborn and Matplotlib", fontsize=16)
ax.set_xlabel("Total Bill ($)", fontsize=12)
ax.set_ylabel("Tip ($)", fontsize=12)

# Add a text annotation using Matplotlib
ax.text(0.95, 0.95, "Matplotlib Text", transform=ax.transAxes, 
        verticalalignment='top', horizontalalignment='right')

# Add a horizontal line using Matplotlib
ax.axhline(y=tips['tip'].mean(), color='r', linestyle='--', label='Average Tip')

# Customize the legend
ax.legend(title="Legend", bbox_to_anchor=(1.05, 1), loc='upper left')

# Adjust the layout and display the plot
plt.tight_layout()
````

This example demonstrates how you can use Seaborn to create the main plot while using Matplotlib functions to add custom elements and adjust the plot's appearance.


### <a id='toc3_3_'></a>[Advanced integration techniques](#toc0_)


1. **Combining different plot types:**


````python
# Create a figure with two y-axes
fig, ax1 = plt.subplots(figsize=(10, 6))
ax2 = ax1.twinx()

# Plot data on the first y-axis using Seaborn
sns.barplot(data=tips, x="day", y="total_bill", ax=ax1, alpha=0.7)
ax1.set_ylabel("Total Bill ($)", color="b")

# Plot data on the second y-axis using Matplotlib
ax2.plot(tips.groupby("day")["tip"].mean(), color="r", marker="o")
ax2.set_ylabel("Average Tip ($)", color="r")

plt.title("Total Bills and Average Tips by Day")
````

2. **Using Seaborn styles with Matplotlib plots:**


````python
# Set Seaborn style
sns.set_style("darkgrid")

# Create a Matplotlib plot
fig, ax = plt.subplots(figsize=(6, 4))
x = np.linspace(0, 10, 100)
ax.plot(x, np.sin(x), label="sin(x)")
ax.plot(x, np.cos(x), label="cos(x)")
ax.set_title("Matplotlib Plot with Seaborn Style")
ax.legend()
plt.show()

# Reset to default style
sns.reset_orig()
````

3. **Customizing Seaborn plots with Matplotlib's object-oriented interface:**


````python
# Create a Seaborn plot
g = sns.FacetGrid(tips, col="time", height=4, aspect=0.8)
g.map(sns.scatterplot, "total_bill", "tip")

# Customize using Matplotlib
for ax in g.axes.flat:
    ax.set_title(ax.get_title().split("=")[1], fontsize=16)
    ax.set_xlabel("Total Bill ($)", fontsize=12)
    ax.set_ylabel("Tip ($)", fontsize=12)

    # Add a diagonal line
    lims = ax.get_xlim()
    ax.plot(lims, lims, ls="--", c=".3", alpha=0.5)
    ax.set_xlim(lims)
    ax.set_ylim(lims)

plt.tight_layout()
````

By understanding how Seaborn integrates with Matplotlib, you can create highly customized visualizations that combine the strengths of both libraries. This integration allows you to use Seaborn's statistical plotting functions and aesthetically pleasing defaults while still having access to Matplotlib's extensive customization options when needed.
